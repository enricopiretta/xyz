```python
def f(ticker):
    data = web.get_data_yahoo(ticker)
    data['Date'] = data.index
    data.index = range(data.shape[0])
    only_friday = data[data.Date.dt.weekday == 4]
    T = only_friday['Date'].to_list()
    dd = (T[-1] - T[0]).days
    S = only_friday['Adj Close'].to_list()
    print(type(S[0]))

    trenda, trendb, a, b = [], [], 0, 0
    for i in range(len(S)-1):
        if S[i] < S[i+1]:
            a += 125
            b += 125
        else:
            a -= 125
            b -= 175
        trenda.append(a)
        trendb.append(b)
    #dd = (only_friday.loc[only_friday.shape[0] -1, 'Date'] - only_friday.loc[0, 'Date']).days
    plt.plot(list(range(len(trenda))), trenda)
    plt.show()
    print('avg_daily_return with 125/125', np.mean(trenda)/dd)
    IRR = (S[-1] - S[0])/dd
    print('avg_daily_return for {} = '.format(ticker), IRR)
    plt.plot(list(range(len(trendb))), trendb)
    plt.show()
    print('avg_daily_return with 125/175', np.mean(trendb)/dd)
```
